```python
import xmitgcm
import xmitgcm.llcreader as llcreader
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.basemap import Basemap, cm, shiftgrid

from dask.distributed import Client, LocalCluster
cluster = LocalCluster(n_workers=3, threads_per_worker=20)
client = Client(cluster)
```

```python
model = llcreader.ECCOPortalLLC2160Model()
```

```python
ds = model.get_dataset(varnames=['U','V','oceTAUX','oceTAUY'], k_levels=[0], type='latlon',
                        read_grid=True)
ds = ds.sel(time=slice("2012-12-01", "2012-12-31"))
ds.time
```

```python
t = ds.time
j = 0
for i in t:
    print(str(ds.time[j].values))
    ti=str(ds.time[j].values)
    ti=ti[0:4]+ti[5:7]+ti[8:10]+"T"+ti[11:13]+ti[14:16]+ti[17:19]    
    ds.U.squeeze().isel(time=[j]).values.astype(">f4")\
    .tofile("/HOME/users/fcampos/outputs/global/U/U_8640x6480_"+ti+".data")
    ds.V.squeeze().isel(time=[j]).values.astype(">f4")\
    .tofile("/HOME/users/fcampos/outputs/global/V/V_8640x6480_"+ti+".data")
    ds.oceTAUX.squeeze().isel(time=[j]).values.astype(">f4")\
    .tofile("/HOME/users/fcampos/outputs/global/TAUX/TAUX_8640x6480_"+ti+".data")
    ds.oceTAUY.squeeze().isel(time=[j]).values.astype(">f4")\
    .tofile("/HOME/users/fcampos/outputs/global/TAUY/TAUY_8640x6480_"+ti+".data")
    j = j+1
```
